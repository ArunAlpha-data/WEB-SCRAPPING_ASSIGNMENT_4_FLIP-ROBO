# 1.To Scrape the details of most viewed videos on YouTube from Wikipedia.

```python
# importing required libraries.
import pandas as pd
import selenium
import time

#importing Webdriver
from selenium import webdriver

#importing Exception Handling
from selenium.common.exceptions import NoSuchElementException
```

```python
#connecting the web driver.
driver=webdriver.Chrome(r'C:\Program Files (x86)\chromedriver_win32\chromedriver.exe')
```

```python
#connecting the website.
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
#creating empty list
Rank=[]
Name=[]
Artist=[]
Upload_Date=[]
Views=[]
```

```python
#creating the list for Rank
rank=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[1]")
for i in rank:
    try:
        Rank.append(i.text.replace(".",""))
    except:
        Rank.append("-")
```

```python
#creating the list for Name
name=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[2]")
for i in name:
    try:
        Name.append(i.text)
    except:
        Name.append("-")
```

```python
#creating the list for Artist
artist=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[3]")
for i in artist:
    try:
        Artist.append(i.text)
    except:
        Artist.append("-")
```

```python
#creating the list for Upload Date
upload=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[5]")
for i in upload:
    try:
        Upload_Date.append(i.text)
    except:
        Upload_Date.append("-")
```

```python
#creating the list for Views
views=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[4]")
for i in views:
    try:
        Views.append(i.text)
    except:
        views.append("-")
```

```python
print(len(Rank),len(Name),len(Artist),len(Upload_Date),len(Views))
```

```python
#creating a Data Frame
Most_Viewed_Youtube_Videos=pd.DataFrame([])
Most_Viewed_Youtube_Videos["Rank"]=Rank
Most_Viewed_Youtube_Videos["Name"]=Name
Most_Viewed_Youtube_Videos["Artist"]=Artist
Most_Viewed_Youtube_Videos["Upload_Date"]=Upload_Date
Most_Viewed_Youtube_Videos["Views"]=Views
```
